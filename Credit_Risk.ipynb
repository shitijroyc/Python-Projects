```python
fully_paid = df[df['loan_status'] == 'Fully Paid']
fully_paid_grade = fully_paid.groupby('grade').size().reset_index(name = 'count')

plt.title('Fully Paid loans with Grades')
sns.barplot(x = 'grade', y = 'count', data = fully_paid_grade, color = 'teal')
plt.show()
```

```python
charged_off = df[df['loan_status'] == 'Charged Off']
charged_off_grade = charged_off.groupby('grade').size().reset_index(name = 'count')

plt.title('Charged off loans with Grades')
sns.barplot(x = 'grade', y = 'count', data = charged_off_grade, color = 'teal')
plt.show()
```

```python
no_policy_off = df[df['loan_status'] == 'Does not meet the credit policy. Status:Charged Off']
no_policy_off_grade = no_policy_off.groupby('grade').size().reset_index(name = 'count')

plt.title('No policy charged off loans with Grades')
sns.barplot(x = 'grade', y = 'count', data = no_policy_off_grade, color = 'teal')
plt.show()
```

We can infer from the above plots that loan risk increases (count for lower grades increases) when the loan status is charged off and no policy charged off.
And loan risk is minimal for fully paid status.

About 70% of fully paid loans have grades between A and C, whereas 70% of charged off loans are between grades B and D.
The condition is much worse for no policy charged off loans.

```python
term_fp = fully_paid.groupby(['term']).size().reset_index(name='count')
term_co = charged_off.groupby(['term']).size().reset_index(name='count') 
term_npco = no_policy_off.groupby(['term']).size().reset_index(name='count')

figure, axis = plt.subplots(3, figsize = (4,8))
axis[0].bar(x = 'term', height = 'count', data = term_fp, color='teal')
axis[0].set_title('Term for fully paid loans')


axis[1].bar(x = 'term', height = 'count', data = term_co, color='teal')
axis[1].set_title('Term for charged off loans')
axis[2].bar(x = 'term', height = 'count', data = term_npco, color='teal')
axis[2].set_title('Term for no policy charged off loans')

plt.subplots_adjust(hspace=1)
plt.show()
```

From the above plot we can infer that short term loan is preffered over long term loan.

```python
fully_paid_purpose = fully_paid.groupby('purpose').size().reset_index(name = 'count')
charged_off_purpose = charged_off.groupby('purpose').size().reset_index(name = 'count')

plt.figure(figsize=(15,5))
sns.barplot(y = 'purpose', x = 'count', data = fully_paid_purpose, color='teal')
plt.title('Purpose for fully paid loans')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(y = 'purpose', x = 'count', data = charged_off_purpose, color='teal')
plt.title('Purpose for charged off loans')
plt.show()
```

```python
fully_paid['purpose'].value_counts(normalize=True).round(3)*100.
```

```python
charged_off['purpose'].value_counts(normalize=True).round(3)*100
```

Debt consolidation and credit card are the top most purposes to apply for a loan.
About 60% of loans are taken for debt consolidation and about 20% for credit card.
Downside to alloting loan for debt is one takes loan to remove another loan which can result in higher interests
and hence increasing the risk for the bank.

```python
fp_int_dc = fully_paid[fully_paid['purpose']=='debt_consolidation']['int_rate']
co_int_dc = charged_off[charged_off['purpose']=='debt_consolidation']['int_rate']
sns.displot(fp_int_dc, kde=True)
plt.title('Fully paid debt consolidation loan')
sns.displot(co_int_dc, kde=True)
plt.title('Charged off debt consolidation loan')
plt.show()
```

From the above plots, we can say that interest rates are somewhat higher for charged off loans than fully paids loans with respect to debt consolidation.
This can be a factor of why people were not able to repay the loan back.

```python
fp_int_cc = fully_paid[fully_paid['purpose']=='credit_card']['int_rate']
co_int_cc = charged_off[charged_off['purpose']=='credit_card']['int_rate']

sns.displot(fp_int_cc, kde=True)
plt.title('Fully paid credit card loan')

sns.displot(co_int_cc,kde=True)
plt.title('Charged off credit card loan')
plt.show()
```

Just like earlier, charged off interest rates are higher than fully paid interest rates.

```python
int_mean_fp = fully_paid.groupby('purpose')['int_rate']
int_mean_fp.mean()
```

```python
int_mean_co = charged_off.groupby('purpose')['int_rate']
int_mean_co.mean()
```

Even the numbers above tell us the same story of why people were charging off of the loan.
They were not able to repay the loan due to higher interest rates.


Recommendations to minimize banking loss/risky consumers:

1 If interests are lower, consumers will not charge off and will try to repay the loan amount.
2 Loan should be sanctioned to those with higher base income.
3 Loan term should be preferrably lower as most consumers apply for short term loans.
4 For the growth of the bank, consumers with higher salaries can be given different interest rate and lesser term
  as they will be able to repay the loan back.
5 Consumers with average income can get the benefit of lesser installments, thus extending the grace period or loan term.
6 Credit cards being the second most generator of loan appeals can have higher interest rates with respect to the consumer's term, loan amount and income.
